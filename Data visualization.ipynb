```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
```

```python
x = np.random.normal(loc=50, scale=10, size=100)  # e.g., age
y = x * 1.5 + np.random.normal(loc=0, scale=15, size=100)  # e.g., salary with noise

# Put in DataFrame
df = pd.DataFrame({'Age': x, 'Salary': y})
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Salary', color='teal')
plt.title('Scatter Plot of Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df[['Age','Salary']],palette='Set2')
plt.title('Box Plot of Age and Salary')
plt.grid(True)
plt.show()
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]
print(outliers)
```

```python
sns.boxplot(y=df['Salary'])
plt.title('Detecting Outliers with Box Plot')
plt.show()
```

```python
sns.scatterplot(x = 'Age', y = 'Salary', data = df)
plt.title('Scatter plot (Outliers are far from the  trend)')
plt.show()
```
